```python
import pandas as pd

pairs = []
for item in vg_objects[:20000]:   # sample subset for faster run
    for obj in item.get("objects", []):
        obj_name = obj.get("names", [""])[0].strip().lower()
        for attr in obj.get("attributes", []):
            pairs.append((attr.lower(), obj_name))

df = pd.DataFrame(pairs, columns=["attribute", "object"])
df = df[df["attribute"].notna() & (df["attribute"] != "")]
df.to_csv("/kaggle/working/attr_obj_pairs.csv", index=False)

print(f"✅ Extracted {len(df)} attribute–object pairs.")
df.head()
```

```python
import pandas as pd

pairs = []
for item in vg_objects[:20000]:  # sample subset for speed
    for obj in item.get("objects", []):
        obj_name = obj.get("names", [""])[0].strip().lower()
        # combine attributes + synsets
        all_attrs = []
        if "attributes" in obj and obj["attributes"]:
            all_attrs.extend(obj["attributes"])
        if "synsets" in obj and obj["synsets"]:
            all_attrs.extend(obj["synsets"])
        for a in all_attrs:
            a = a.replace(".n.01", "").replace(".v.01", "").strip().lower()
            if a and obj_name:
                pairs.append((a, obj_name))

df = pd.DataFrame(pairs, columns=["attribute", "object"])
df = df.dropna().query("attribute != '' and object != ''")
df.to_csv("/kaggle/working/attr_obj_pairs.csv", index=False)

print(f"✅ Extracted {len(df)} attribute–object pairs.")
df.head(10)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import numpy as np

# Directory of labels
label_dir = "/kaggle/input/coco128/coco128/labels/train2017"

# Collect all class IDs
all_classes = []
for file in os.listdir(label_dir):
    df = pd.read_csv(os.path.join(label_dir, file), sep=' ', header=None)
    all_classes.extend(df[0].tolist())

# Create frequency dataframe
df_all = pd.DataFrame(all_classes, columns=["class_id"])
class_counts = df_all["class_id"].value_counts().sort_index()
class_counts_df = pd.DataFrame({"class_id": class_counts.index, "count": class_counts.values})

print("Total unique classes:", len(class_counts))
print(class_counts_df.head())
```

```python
plt.figure(figsize=(7,7))
plt.pie(class_counts_df["count"].head(10),
        labels=class_counts_df["class_id"].head(10),
        autopct='%1.1f%%', startangle=90, colors=plt.cm.tab10.colors)
plt.title("Top 10 Object Classes (COCO128)", fontsize=14)
plt.tight_layout()
plt.savefig("/kaggle/working/fig_coco_pie_full.png", dpi=200)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(class_counts_df["class_id"], class_counts_df["count"], color='royalblue', marker='o')
plt.title("COCO128 Class Frequency Distribution", fontsize=14)
plt.xlabel("Class ID")
plt.ylabel("Count")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("/kaggle/working/fig_coco_line_full.png", dpi=200)
plt.show()
```

```python
# Build co-occurrence matrix for top classes
top_classes = class_counts_df["class_id"].head(15).tolist()
co_matrix = pd.DataFrame(0, index=top_classes, columns=top_classes)

for file in os.listdir(label_dir):
    df = pd.read_csv(os.path.join(label_dir, file), sep=' ', header=None)
    ids = df[0].tolist()
    for a in ids:
        for b in ids:
            if a in top_classes and b in top_classes and a != b:
                co_matrix.loc[a, b] += 1

plt.figure(figsize=(8,6))
sns.heatmap(co_matrix, cmap="coolwarm", annot=True, fmt="d")
plt.title("Top 15 Class Co-occurrence Heatmap (COCO128)", fontsize=13)
plt.tight_layout()
plt.savefig("/kaggle/working/fig_coco_heatmap_full.png", dpi=200)
plt.show()
```

```python
from PIL import Image
import pandas as pd
import os

img_data = []
sample_files = [f for f in os.listdir(IMG_DIR) if f.endswith('.jpg')][:500]  # sample 500 to save time

for img_name in sample_files:
    path = os.path.join(IMG_DIR, img_name)
    with Image.open(path) as img:
        width, height = img.size
    img_data.append((img_name, width, height, os.path.getsize(path)/1024))  # size in KB

df = pd.DataFrame(img_data, columns=["filename", "width", "height", "size_kb"])
df["aspect_ratio"] = round(df["width"] / df["height"], 2)
df.describe()
```

```python
import matplotlib.pyplot as plt

ratio_counts = df["aspect_ratio"].round().value_counts().head(5)

plt.figure(figsize=(6,6))
plt.pie(ratio_counts, labels=ratio_counts.index, autopct="%1.1f%%", startangle=90, colors=plt.cm.Paired.colors)
plt.title("Top 5 Image Aspect Ratios (Flickr30k)", fontsize=13)
plt.tight_layout()
plt.savefig("/kaggle/working/fig_flickr_pie_aspect_ratios.png", dpi=200)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["width"], df["height"], 'o', alpha=0.5)
plt.title("Image Width vs Height (Flickr30k Sample)", fontsize=13)
plt.xlabel("Width (px)")
plt.ylabel("Height (px)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("/kaggle/working/fig_flickr_line_dimensions.png", dpi=200)
plt.show()
```

```python
import os, glob
from PIL import Image
import pandas as pd
import numpy as np
from tqdm import tqdm

# YOLO-style COCO128 paths
label_dir = "/kaggle/input/coco128/coco128/labels/train2017"
image_dir = "/kaggle/input/coco128/coco128/images/train2017"

print("Labels path:", label_dir)
print("Images path:", image_dir)
print("Num label files:", len(glob.glob(os.path.join(label_dir, "*.txt"))))
```

**Cell 2 – COCO80 class names (for mapping ids → names)**

```python
# Standard COCO 80 class names (YOLO order)
coco_classes = [
    'person','bicycle','car','motorcycle','airplane','bus','train','truck','boat',
    'traffic light','fire hydrant','stop sign','parking meter','bench',
    'bird','cat','dog','horse','sheep','cow','elephant','bear','zebra','giraffe',
    'backpack','umbrella','handbag','tie','suitcase','frisbee','skis','snowboard',
    'sports ball','kite','baseball bat','baseball glove','skateboard','surfboard',
    'tennis racket','bottle','wine glass','cup','fork','knife','spoon','bowl',
    'banana','apple','sandwich','orange','broccoli','carrot','hot dog','pizza',
    'donut','cake','chair','couch','potted plant','bed','dining table','toilet',
    'tv','laptop','mouse','remote','keyboard','cell phone','microwave','oven',
    'toaster','sink','refrigerator','book','clock','vase','scissors','teddy bear',
    'hair drier','toothbrush'
]

len(coco_classes)
```

**STEP 4B – Load YOLO labels and convert to pixel boxes**

**Cell 3 – Helper to convert YOLO → pixel x1,y1,x2,y2**

```python
def yolo_to_xyxy(img_path, xc, yc, w, h):
    img = Image.open(img_path).convert("RGB")
    W, H = img.size
    x1 = (xc - w/2) * W
    y1 = (yc - h/2) * H
    x2 = (xc + w/2) * W
    y2 = (yc + h/2) * H
    return img, (int(x1), int(y1), int(x2), int(y2))
```

**Cell 4 – Build a dataframe of all GT boxes**

```python
annotations = []

label_files = sorted(glob.glob(os.path.join(label_dir, "*.txt")))

for lf in tqdm(label_files, desc="Reading YOLO labels"):
    base = os.path.basename(lf).replace(".txt", "")
    img_path = os.path.join(image_dir, base + ".jpg")
    if not os.path.exists(img_path):
        img_path = os.path.join(image_dir, base + ".png")
        if not os.path.exists(img_path):
            # skip if no image
            continue

    with open(lf, "r") as f:
        lines = f.read().strip().split("\n")

    for line in lines:
        if not line:
            continue
        cls_id, xc, yc, w, h = map(float, line.split())
        cls_id = int(cls_id)
        if cls_id < 0 or cls_id >= len(coco_classes):
            continue
        cat_name = coco_classes[cls_id]
        img, (x1, y1, x2, y2) = yolo_to_xyxy(img_path, xc, yc, w, h)
        annotations.append({
            "image_path": img_path,
            "cls_id": cls_id,
            "gt_cat": cat_name,
            "bbox": (x1, y1, x2, y2)
        })

len(annotations)
```

**Cell 5 – Put into a DataFrame**

```python
dfb = pd.DataFrame(annotations)
dfb.head()
```

**STEP 4C — CATEGORY-ONLY BASELINE**

**Cell 6 — Build category-only CLIP text embeddings**

```python
import torch.nn.functional as F

@torch.no_grad()
def build_category_prompts(categories, tokenizer, model):
    prompts = [f"a photo of a {c}" for c in categories]
    tok = tokenizer(prompts).to(device)
    text_feats = model.encode_text(tok)
    text_feats = F.normalize(text_feats, dim=-1)
    return text_feats

# build text embeddings for all 80 COCO classes
cat_text_feats = build_category_prompts(coco_classes, tokenizer, model)
cat_text_feats.shape
```

**Cell 7 — Function to crop region & extract CLIP image embedding**

```python
@torch.no_grad()
def encode_crop(img_path, bbox):
    x1, y1, x2, y2 = bbox
    img = Image.open(img_path).convert("RGB")
    crop = img.crop((x1, y1, x2, y2))
    crop_tensor = preprocess(crop).unsqueeze(0).to(device)
    img_feat = model.encode_image(crop_tensor)
    img_feat = F.normalize(img_feat, dim=-1)
    return img_feat
```

**Cell 8 — Predict categories using CLIP similarity**


```python
preds = []

for idx, row in tqdm(dfb.iterrows(), total=len(dfb), desc="Running CLIP baseline"):
    img_feat = encode_crop(row["image_path"], row["bbox"])
    sims = (img_feat @ cat_text_feats.T).squeeze(0)
    pred_id = int(torch.argmax(sims).item())
    preds.append(pred_id)

dfb["pred_id"] = preds
dfb["pred_cat"] = dfb["pred_id"].apply(lambda i: coco_classes[i])
dfb.head()
```

**Cell 9 — Compute baseline metrics**


```python
# Overall accuracy
overall_acc = (dfb.gt_cat == dfb.pred_cat).mean()

seen_cats = coco_classes[:60]
unseen_cats = coco_classes[60:]

seen_mask = dfb.gt_cat.isin(seen_cats)
unseen_mask = dfb.gt_cat.isin(unseen_cats)

seen_acc = (dfb[seen_mask].gt_cat == dfb[seen_mask].pred_cat).mean()
unseen_acc = (dfb[unseen_mask].gt_cat == dfb[unseen_mask].pred_cat).mean()

def harmonic_mean(a, b, eps=1e-8):
    return 2 * a * b / (a + b + eps)

hm = harmonic_mean(seen_acc, unseen_acc)

dfb.to_csv("baseline_results.csv", index=False)

overall_acc, seen_acc, unseen_acc, hm
```

**Step 4C — Build Attribute prompts**

```python
def build_attr_prompts(categories, attributes, tokenizer, model, batch_size=32):
    prompts = []
    for cat in categories:
        for attr in attributes:
            prompts.append(f"a {attr} {cat}")

    all_feats = []

    for i in range(0, len(prompts), batch_size):
        batch_prompts = prompts[i:i+batch_size]
        tok = tokenizer(batch_prompts).to(device)

        with torch.no_grad():
            feats = model.encode_text(tok)
            feats = feats / feats.norm(dim=-1, keepdim=True)

        all_feats.append(feats.cpu())

    all_feats = torch.cat(all_feats, dim=0)
    return prompts, all_feats
```

```python
# ======================================================
# STEP 4E — Reasoning + Confidence + Calibration (ECE)
# ======================================================

import torch
import numpy as np
import pandas as pd

# ------------------------------------------------------
# 1. SIMPLE KG REASONING FILTER (NEURO-SYMBOLIC RULES)
# ------------------------------------------------------

invalid_pairs = {
    ("metal", "cat"),
    ("metal", "dog"),
    ("plastic", "sandwich"),
    ("wooden", "pizza"),
    ("square", "person"),
    ("round", "car"),
}

def kg_rule(attr, cat):
    if (attr, cat) in invalid_pairs:
        return "invalid"
    return "valid"

dfb["kg_flag"] = dfb.apply(lambda r: kg_rule(r["pred_attr"], r["pred_cat_attr"]), axis=1)


# ------------------------------------------------------
# 2. EXTRACT CONFIDENCE FROM 4D PREDICTIONS
# ------------------------------------------------------

all_conf = []

device = "cuda" if torch.cuda.is_available() else "cpu"

# move attribute text features to device (safety)
attr_text_feats = attr_text_feats.to(device)

for idx, row in dfb.iterrows():

    # encode image region: (YOUR FUNCTION FROM STEP 4D)
    img_feat = encode_crop(row["image_path"], row["bbox"]).to(device)

    # cosine similarity with all prompts
    sims = (img_feat @ attr_text_feats.T)[0]

    # take the max score as confidence
    confidence = float(torch.max(sims))
    all_conf.append(confidence)

dfb["attr_conf"] = all_conf


# ------------------------------------------------------
# 3. NORMALIZE CONFIDENCE TO [0,1]
# ------------------------------------------------------

dfb["attr_conf_norm"] = (
    (dfb["attr_conf"] - dfb["attr_conf"].min()) /
    (dfb["attr_conf"].max() - dfb["attr_conf"].min() + 1e-8)
)


# ------------------------------------------------------
# 4. COMPUTE ECE (Expected Calibration Error)
# ------------------------------------------------------

def compute_ece(pred_correct, conf, n_bins=10):
    bins = np.linspace(0, 1, n_bins+1)
    ece = 0.0
    
    for i in range(n_bins):
        start, end = bins[i], bins[i+1]
        mask = (conf >= start) & (conf < end)
        
        if np.sum(mask) > 0:
            acc = np.mean(pred_correct[mask])
            avg_conf = np.mean(conf[mask])
            ece += (np.sum(mask) / len(conf)) * abs(acc - avg_conf)
    
    return ece

# 1 = correct, 0 = incorrect
pred_correct = (dfb["gt_cat"] == dfb["pred_cat_attr"]).astype(int).values
conf = dfb["attr_conf_norm"].values

ece_attr = compute_ece(pred_correct, conf)

print("ECE (Attribute-aware model):", ece_attr)


# ------------------------------------------------------
# 5. SAVE RESULTS (OPTIONAL)
# ------------------------------------------------------

dfb.to_csv("step4_attribute_model_outputs.csv", index=False)
print("Saved:", "step4_attribute_model_outputs.csv")
```

## 4. Model Design

The proposed zero-shot object detection framework follows a modular design that integrates vision–language models, compositional prompts, and lightweight knowledge-based reasoning. The model consists of five major components: (1) image–text embedding backbone, (2) region proposal module, (3) compositional text encoder, (4) similarity-based detection head, and (5) logical reasoning and calibration layers.

### 4.1 Backbone: CLIP ViT-B/16
CLIP (Contrastive Language–Image Pretraining) serves as the foundation of the system. The ViT-B/16 variant is used to embed visual regions and textual prompts into a shared semantic space. This enables zero-shot transfer without task-specific training.  
- **Image Encoder:** Produces a 512-dim feature vector for each region.  
- **Text Encoder:** Maps textual prompts (category-only or attribute-aware) into the same embedding space.

### 4.2 Region Proposal Module
To localize candidate objects, the framework relies on CLIP-guided region cropping. Although the full system can integrate Grounding DINO or Faster R-CNN RPN, the current implementation uses CLIP similarity scoring over pre-defined crops.  
- Efficient, training-free region extraction.  
- Suitable for zero-shot scenarios without fine-tuning.

### 4.3 Attribute-aware Text Encoder
Instead of relying solely on category names (e.g., “zebra”), the model constructs **compositional prompts** that combine:  
- Object category (e.g., *zebra*)  
- Visual attributes (e.g., *striped, black-and-white, animal*)  
- KG-derived relations (e.g., *has_texture(striped)*)

Example prompt:  
**“a striped black-and-white animal (zebra)”**

The text encoder converts these enriched prompts into attribute-aware embeddings that improve generalization to unseen categories.

### 4.4 Similarity Matching (Cosine Similarity Head)
For each region proposal, cosine similarity is computed against all prompt embeddings:  

\[
\text{score}(r, t) = \frac{f_\text{img}(r) \cdot f_\text{text}(t)}{\|f_\text{img}(r)\| \, \|f_\text{text}(t)\|}
\]

Regions are classified as the category with the highest similarity score.

### 4.5 Logical Reasoning Constraints (Neuro-symbolic Layer)
To improve interpretability and avoid inconsistent predictions, the model incorporates light-weight reasoning rules:

- **Invalid attribute–category pairs:**  
  Examples: (“metal”, “cat”), (“round”, “car”), (“plastic”, “sandwich”)

- **Consistency rule:**  
  *“striped AND animal → NOT vehicle”*

These rules filter out logically impossible predictions, improving robustness and interpretability.

### 4.6 Knowledge Graph Integration
A small attribute–object graph is constructed from Visual Genome. The graph encodes relations such as:
- **has_color(x, red)**  
- **has_texture(x, striped)**  
- **has_shape(x, round)**

Graph-derived attribute relations help enrich the prompts and reinforce consistency with real-world semantics.

### 4.7 Calibration Layer
To assess uncertainty and trustworthiness:
- Confidence distributions are normalized.
- Expected Calibration Error (ECE) is computed.
- Reliability diagrams compare predicted confidence vs actual accuracy.

This step ensures that the model’s predictions can be trusted in zero-shot settings, where uncertainty plays a major role.

### Summary
The final architecture integrates:
- Vision–language embeddings (CLIP)  
- Compositional attribute-aware prompts  
- Region proposal and similarity matching  
- Knowledge-graph reasoning  
- Uncertainty calibration  

Together, these components form a knowledge-driven, interpretable zero-shot detection system.

**STEP 5 — Cell #4: Compute accuracies + harmonic mean**

```python
# STEP 5 — Compute accuracy for baseline vs attribute-aware

def compute_acc(df, pred_col, seen_classes, unseen_classes):
    gt = df["gt_cat"]

    overall_acc = (gt == df[pred_col]).mean()

    seen_mask = gt.isin(seen_classes)
    unseen_mask = gt.isin(unseen_classes)

    seen_acc = (gt[seen_mask] == df.loc[seen_mask, pred_col]).mean()
    unseen_acc = (gt[unseen_mask] == df.loc[unseen_mask, pred_col]).mean()

    # harmonic mean of seen/unseen
    if seen_acc + unseen_acc > 0:
        hm = 2 * seen_acc * unseen_acc / (seen_acc + unseen_acc)
    else:
        hm = 0.0

    return overall_acc, seen_acc, unseen_acc, hm


# Baseline: use 'pred_cat'
base_overall, base_seen, base_unseen, base_hm = compute_acc(
    df_base, "pred_cat", seen_classes, unseen_classes
)

# Attribute-aware: use 'pred_cat_attr'
attr_overall, attr_seen, attr_unseen, attr_hm = compute_acc(
    df_attr, "pred_cat_attr", seen_classes, unseen_classes
)

print("Baseline model:")
print("  Overall acc :", round(base_overall, 4))
print("  Seen acc    :", round(base_seen, 4))
print("  Unseen acc  :", round(base_unseen, 4))
print("  HM          :", round(base_hm, 4))

print("\nAttribute-aware model:")
print("  Overall acc :", round(attr_overall, 4))
print("  Seen acc    :", round(attr_seen, 4))
print("  Unseen acc  :", round(attr_unseen, 4))
print("  HM          :", round(attr_hm, 4))
```

### Architecture Overview

The proposed framework follows a modular, knowledge-driven architecture optimized for zero-shot object detection. The pipeline begins with region proposal generation from the input image, followed by CLIP’s ViT-B/16 encoder extracting semantic visual embeddings for each region.

Parallel to this, the system constructs compositional prompts that combine category names with visual attributes (color, texture, shape) and knowledge-graph relations. These enriched prompts are encoded using CLIP’s text encoder.

A cosine similarity head matches region embeddings to prompt embeddings, producing zero-shot class scores without any fine-tuning. A lightweight neuro-symbolic reasoning layer filters logically inconsistent predictions using invalid attribute–category rules.

This architecture enables interpretable zero-shot transfer, improved unseen-category recognition, and stable cross-dataset generalization.
